Based on the dataset, possible research questions could be:

## Data Summary

The data used for the current project is a publicly available **High School Directory** Historical data provided by the **Department of Education (DOE) NY** [1].

The data set provides information on 437 use cases (i.e., rows) where each row represents a unique school in New York City, Which including the school's name, location, grade span, extracurricular activities, and partnerships with external organizations. Each row in the dataset represents a high school with 67 attributes including:

- dbn: Unique identifier for each school (string)
- school_name: Name of the school (string)
- borough: Borough where the school is located (string)
- building_code: Building code for the school (string)
- phone_number: Phone number for the school (string)
- fax_number: Fax number for the school (string)
- grade_span_min: Minimum grade offered by the school (integer)
- grade_span_max: Maximum grade offered by the school (integer)
- expgrade_span_min: Minimum expected grade offered by the school (empty or integer)
- expgrade_span_max: Maximum expected grade offered by the school (empty or integer)
- bus: Bus routes that stop near the school (string)
- subway: Subway lines that stop near the school (string)
- primary_address_line_1: Street address of the school (string)
- city: City where the school is located (string)
- state_code: State where the school is located (string)
- postcode: ZIP code where the school is located (string)
- website: Website of the school (string)
- total_students: Total number of students in the school (integer)
- campus_name: Name of the campus where the school is located (empty or string)
- school_type: Type of the school (string)
- overview_paragraph: Overview of the school (string)
- program_highlights: Highlights of the school's programs (string)
- language_classes: Language classes offered by the school (string)
- advancedplacement_courses: Advanced Placement courses offered by the school (string)
- online_ap_courses: Online Advanced Placement courses offered by the school (string)
- online_language_courses: Online language courses offered by the school (string)
- extracurricular_activities: Extracurricular activities offered by the school (string)
- psal_sports_boys: PSAL sports offered for boys (string)
- psal_sports_girls: PSAL sports offered for girls (string)
- psal_sports_coed: PSAL sports offered for both boys and girls (string)
- school_sports: Other sports offered by the school (string)
- partner_cbo: Community-based organizations that partner with the school (string)
- partner_hospital: Hospitals that partner with the school (string)
- partner_highered: Higher education institutions that partner with the school (string)
- partner_cultural: Cultural institutions that partner with the school (string)
- partner_nonprofit: Nonprofit organizations that partner with the school (string)
- partner_corporate: Corporate partners of the school (string)
- partner_financial: Financial partners of the school (string)
- partner_other: Other partners of the school (string)
- addtl_info1: Additional information about the school (string)
- addtl_info2: Additional information about the school (string)
- start_time: Start time for the school day (string)
- end_time: End time for the school day (string)
- se_services: Special education services offered by the school (string)
- ell_programs: English Language Learner programs offered by the school (string)
- school_accessibility_description: Accessibility description for the school (string)
- number_programs: Number of programs offered by the school (integer)
- priority01 to priority10: Admissions priorities for the school (string)
- Location 1: Geographical location of the school in latitude (string)
- Location 1: Geographical location of the school in longitude (string)
- Community Board: Community Board where the school is located (integer)
- Council District: Council District where the school is located (integer)
- Census Tract: Census Tract where the school is located (integer)
- Bin : building identifaction number of the school (integer)
- BBL: borough block and lot of school (Integer)
- NTA: neighborhood tabulation area (String)

Manhattan schools tend to offer the most extracurricular activities.

Other observations:

Staten Island schools tend to offer the most extracurricular activities with 30 or more offerings.
Schools in all boroughs offer a relatively similar number of extracurricular activities with 1-9 offerings.
Overall, the heatmap suggests that there is a relatively equitable distribution of extracurricular activities across boroughs.

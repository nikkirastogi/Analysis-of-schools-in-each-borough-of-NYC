## 01 Introduction

### Brief Summary
The dataset contains details of various schools in **New York City**, including their names, borough, phone number, website, total number of students, academic programs, extracurricular activities, partnerships with external organizations, and more.

### Research question
Based on the dataset, a possible research question could be:

***"Which schools in New York City offer the most extensive extracurricular activities and partnerships with external organizations, and how does this correlate with their student enrollment?"***

The proposed research question intends to examine this dataset to find schools that have the greatest extracurricular offerings and collaborations with other groups. It also aims to look at the relationship between these elements and student enrollment at the schools. 
Parents and educators who want to find schools that give their pupils opportunity for a well-rounded education and experiences may find this information valuable.

## 02 Data Summary

The data used for the current project is a publicly available **High School Directory** Historical data provided by the **Department of Education (DOE) NY** [1].

The data set provides information on 437 use cases (i.e., rows) where each row represents a unique school in New York City, Which including the school's name, location, grade span, extracurricular activities, and partnerships with external organizations. ## Inference

To investigate the realtionship between Number of students vs Number of Extracurricular Activities and Number of Partner Opportunities, we can use Heatmaps to observer the correlation bettwen those variables.

```python
schools_corr = schools[["total_students", "num_ext_act", "num_sports_tot", "num_prtn_tot"]].corr()

sb.heatmap(schools_corr, annot=True)
plt.title("Correlation between Total Students, Number of Extracurricular Activities, Sports Offered, and Partner Opportunities")
plt.show()
```

##### Observation 01
> There is a moderate correlation of 0.59 between Total Number of students and Number of Extracurricular Activities.
> Also 0.53 moderate correlation between Total Number of students and Number of Sports Offered.
> But the relation between number of total sprots students enrollement and number of total Partner Opportunities is less correlated as 0.089.


```python
schools_corr = schools[["total_students", "num_sports_boys", "num_sports_girls", "num_sports_coed", "num_sports_othr", "num_sports_tot"]].corr()

sb.heatmap(schools_corr, annot=True)
plt.title("Heatmap of Total Number of Students and Sports Offered")
plt.show()
```
